### Encode the data

Four columns (`satisfaction`, `customer_type`, `type_of_travel`, `class`) are the pandas dtype object. Decision trees need numeric columns. We must convert these columns to a numeric dtypes if we want to use them in our model. There are a number of ways to do this but for features with two classes we can use `np.where()` method and use `pd.get_dummies()` method for features with more than two classes.

```python
# encode data 
# satisfaction
airline_df['satisfaction'] = np.where(airline_df['satisfaction'] == 'satisfied', 1, 0).astype(np.int64)

# customer_type
airline_df['customer_type'] = np.where(airline_df['customer_type'] == 'Loyal Customer', 1, 0).astype(np.int64)

# type_of_travel
airline_df['type_of_travel'] = np.where(airline_df['type_of_travel'] == 'Business Travel', 1, 0).astype(np.int64)

# class
airline_df = pd.get_dummies(airline_df, columns = ['class'], drop_first = True, dtype = np.int64)

#confirm_changes
airline_df.head(10)                       
```

also check dtypes for new columns

```python
airline_df.dtypes
```

### Create the training and testing data

Put 75% of the data into a training set and the remaining 25% into a testing set. 

```python
y = airline_df['satisfaction']
x = airline_df.drop('satisfaction', axis = 1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)
```

## Step 3: Model building

Fit a decision tree classifier model to the data

```python
# instantiate decision tree classifier
decision_tree = DecisionTreeClassifier(random_state = 0)

# fit model 
dt = decision_tree.fit(x_train, y_train)

# get predictions
dt_pred = dt.predict(x_test)
```

## Step 4: Results and evaluation

Print out the decision tree model's accuracy, precision, recall, and F1 score.

```python
# display evaluation metrices
accuracy = round(accuracy_score(y_test, dt_pred) * 100, 2)
precision = round(precision_score(y_test, dt_pred) * 100, 2)
recall = round(recall_score(y_test, dt_pred) * 100, 2)
f1 = round(f1_score(y_test, dt_pred) * 100, 2)

print( f' Accuracy:  {accuracy}%')
print( f' Precision: {precision}%')
print( f' Recall:    {recall}%')
print( f' F1-Score:  {f1}%')
```

### Plot a confusion matrix

To know the types of errors made by our algorithm, we will plot a confusion matrix.

```python
# confusion matrix
cm = confusion_matrix(y_test, dt_pred, labels = dt.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = dt.classes_)
disp.plot()
plt.show()
```

- The confusion matrix tree shows very high proportions for both true positives and true negatives with both values above 13000 
- The matrix also shows a lower proportion for both false positive and false negative with predictions resulting in a false positive or false negative only 8% of the time, which is a considerably low fail rate.

  This explains the high scores for accuracy, precision and recall

### Plot the decision tree

We will plot a visual representation of the tree to pinpoint where the splits in the data are occuring using `plot_tree()`

```python
plt.figure(figsize=(15, 10))
plot_tree(decision_tree, max_depth=2, fontsize=14, feature_names = x.columns);
```

### Calculate feature importance

The `DecisionTreeClassifier` class comes with a `feature_importances_` attribute that provides access to the features' relative importance in the model. The feature importances are computed when the `fit()` method is called on the `DecisionTreeClassifier` instance.

```python
# calculate feature importance and display plot

importances = decision_tree.feature_importances_
forest_importances = pd.Series(importances, index=x.columns).sort_values()

plt.figure(figsize = (7,5))
forest_importances.plot.barh()
plt.title('Feature Importance Plot')
plt.ylabel('Features')
plt.xlabel('Importance Score')
plt.show()
```

The plot above indicates what features were most important to customer satisfaction in our model. 

Inflight_entertainment was by far the most important feature to customer satisfaction.

The top five features in our model in order of importance to customer satisfaction are:
- inflight_entertainment
- seat_comfort
- ease_of_online_booking
- flight_distance
- depature/arrival_time_convinient
